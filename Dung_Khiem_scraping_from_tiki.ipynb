```
# import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```
def scrape_tiki(url):
  """ This function will get data from tiki website
  """
  # get HTML data from tiki
  try:
    data_html = requests.get(url)
  except Exception as err:
    print('There was a problem: {}'.format(err))
  
  # parse HTML data into a beautiful soup
  soup = BeautifulSoup(data_html.text)
  
  # get a list of items using BeautifulSoup function
  tiki_list = soup.find_all("div", class_= 'product-item')
  
  # get name, price, seller_id, data_id, category, image put into dictionary "d" then for each d, we put them in "product_list"
  product_list = []
  for item in tiki_list:
    d = {}
    d['name'] = item['data-title']
    d['price'] = item['data-price']
    d['seller_id'] = item['data-seller-product-id']
    d['data-id'] = item['data-id']
    d['category'] = item['data-category']
    d['img'] = item.img['src']
    product_list.append(d)
    
  # put product list into pandas dataframe
  df = pd.DataFrame(data=product_list)
  return df
  
scrape_tiki('https://tiki.vn/dien-thoai-may-tinh-bang/c1789?src=c.1789.hamburger_menu_fly_out_banner')
```
